```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as PillowWriter
import timeit
from IPython import display
```

```python
start = timeit.default_timer()
dt = 0.01
T = 100

t = np.arange(0,T,dt)
N = len(t)
x,v,y,w = np.zeros(N),np.zeros(N),np.zeros(N),np.zeros(N),
x[0] = 1
y[0] = 0
v[0] = 0
w[0] = 0.5
```

```python
for i in range(0,N-1):
    t1 = (x[i]**2)+(y[i]**2)
    a1 = -(x[i])/(t1**(3/2))
    b1 = -(y[i])/(t1**(3/2))
    v[i+1] = v[i]+a1*dt
    w[i+1] = w[i]+b1*dt
    x[i+1] = x[i]+v[i+1]*dt
    y[i+1] = y[i]+w[i+1]*dt
```

```python
fig, ax = plt.subplots()
line, = ax.plot(x, y)   
plt.scatter(0,0,color='k')
plt.title('Semi-Implicit Eulers Method for Simple Orbit')
```
